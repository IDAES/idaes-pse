```python
from idaes.dmf import codesearch as cs
import idaes

walker = cs.ModuleClassWalker(from_pkg=idaes, parent_class=idaes.core.base.unit_model.UnitModelBlockData)
```

```python
visitor = cs.Visitor()
walker.walk(visitor)
print("name,module,description")
rows = []
for clazz in visitor.saved:
    name = clazz.__name__
    doc = clazz.__doc__ or "-"
    module = clazz.__module__
    period = doc.find(".")
    if period == -1:
        period = len(doc)
    args = doc.find("Args:")
    if args == -1:
        args = len(doc)
    pos = min(period, args)
    if pos > 0:
        desc = doc[:pos].strip()
    else:
        desc = doc.strip()
    if len(desc) == 0:
        desc = "-"
    desc = desc.replace("\n", " ")
    rows.append(f'{name},{module},"{desc}"')

rows.sort()
print("\n".join(rows))
```
